# Health Monitoring dataset.

```python
# Importing Lib
import pandas as pd
import os

# Load the dataset
file_path = "D:\Machine_Learning_Project\AI_for_Elderly_Care_and_Support\Dataset\health_monitoring.csv"
df = pd.read_csv(file_path)

# Display basic info and first few rows
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
# Checking Null
df.isnull().sum()
```

```python
# Converting the Timestamp to datetime
df["Timestamp"] = pd.to_datetime(df["Timestamp"],errors="coerce")

```

```python
# Spliting 'Blood Pressure' into 'Systolic_BP' and 'Diastolic_BP'
df[['Systolic_BP', 'Diastolic_BP']] = df['Blood Pressure'].str.extract(r'(\d+)/(\d+)').astype(float)
```

```python
#Converting 'Yes/No' columns to binary (1 for 'Yes', 0 for 'No')
yes_no_coumns= [
    'Heart Rate Below/Above Threshold (Yes/No)',
    'Blood Pressure Below/Above Threshold (Yes/No)',
    'Glucose Levels Below/Above Threshold (Yes/No)',
    'SpO₂ Below Threshold (Yes/No)',
    'Alert Triggered (Yes/No)',
    'Caregiver Notified (Yes/No)'
]
```

```python
df[yes_no_coumns] = df[yes_no_coumns].applymap(lambda x: 1 if x.strip().lower()=="yes" else 0)
```

```python
# Renaming columns for consistency
df = df.rename(columns={
    'Device-ID/User-ID': 'user_id',
    'Timestamp': 'timestamp',
    'Heart Rate': 'heart_rate',
    'Heart Rate Below/Above Threshold (Yes/No)': 'hr_alert',
    'Blood Pressure Below/Above Threshold (Yes/No)': 'bp_alert',
    'Glucose Levels': 'glucose_level',
    'Glucose Levels Below/Above Threshold (Yes/No)': 'glucose_alert',
    'Oxygen Saturation (SpO₂%)': 'spo2',
    'SpO₂ Below Threshold (Yes/No)': 'spo2_alert',
    'Alert Triggered (Yes/No)': 'alert_triggered',
    'Caregiver Notified (Yes/No)': 'caregiver_notified',
    'Systolic_BP': 'systolic_bp',
    'Diastolic_BP': 'diastolic_bp'
})
```

```python
# Droping original 'Blood Pressure' column (since we've split it)
df.drop(columns=["Blood Pressure"], inplace=True)
```

```python
# Saving the cleaned data to cleaned data folder
data_folder = "../cleaned_data"
file_name = "cleaned_health_monitoring_dataset.csv"
file_path = os.path.join(data_folder, file_name)
df.to_csv(file_path, index=False)
```

## Connecting to the postgreSQL

```python
import psycopg2
import pandas as pd
```

```python
# Connecting to the sql
conn = psycopg2.connect(
    dbname = "elderly_care",
    user = "postgres",
    password = "system",
    host = "localhost",
    port= "5432"   
    
)

cur = conn.cursor()
```

```python
conn.commit()
cur.close()
conn.close()
```
